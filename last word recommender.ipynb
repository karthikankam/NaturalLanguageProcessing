```python
import pandas as pd
data=pd.read_csv('lastword/sample_reuters_dataset.csv')
data
```

```python
def clean(t):
    #t=re.sub('\(\.*?\)',' ',t)
    t=re.sub('[^a-zA-Z\']'," ",t)
    t=t.lower()
    return t
```

```python
data['cleaned']=data['sentence_text'].apply(clean)
```

```python
vocab=list()
for i in data.cleaned:
    for j in i.split():
        vocab.append(j)
woco=vocab.copy()
vocab=set(vocab)
```

```python
len(vocab)
```

```python
diwo=dict.fromkeys(vocab,0)
for i in data.cleaned:
    for j in i.split():
        diwo[j]+=1
diwo
```

```python
df=pd.DataFrame({'word':list(diwo.keys()),'freq':list(diwo.values())})
```

```python
df
```

```python
def trigram(t):
    tri=[]
    y=t.split()
    for i in range(len(t.split())-2):
        tri.append(y[i:i+3])
    return tri
```

```python
data['tri']=data['cleaned'].apply(trigram)
```

```python
from collections import Counter,defaultdict
```

```python
model=defaultdict(lambda: defaultdict(lambda: 0))
print(model)
for i in range(len(data)):
    for w1,w2,w3 in data['tri'][i]:
        model[(w1,w2)][w3]+=1
```

```python
model
```

```python
def bigram(t):
    bi=[]
    u=t.split()
    for i in range(len(u)-1):
        bi.append(u[i:i+2])
    return bi
```

```python
data['bi']=data['cleaned'].apply(bigram)
```

```python
data
```

```python
model1=defaultdict(lambda: defaultdict(lambda: 0))
for i in range(len(data)):
    for w1,w2 in data['bi'][i]:
        model1[w1][w2]+=1
```

```python
for w1,w2 in model:
    tot=float(sum(model[w1,w2].values()))
    for w3 in model[w1,w2]:
        model[w1,w2][w3]/=tot  
```

```python
for w1 in model1.keys():
    tot1=float(sum(model1[w1].values()))
    for w2 in model1[w1]:
        model1[w1][w2]/=tot1
```

```python
print(model1.keys())
r=input('enter a word')
d=pd.DataFrame({'word':list(model1[r].keys()),'probability':list(model1[r].values())})
d.sort_values('probability',ascending=False,inplace=True)
d.reset_index(drop=True,inplace=True)
print(d)
r1=input('enter the next word')
d=pd.DataFrame({'word':list(model[r,r1].keys()),'probability':list(model[r,r1].values())})
d.sort_values('probability',ascending=False,inplace=True)
d.reset_index(drop=True,inplace=True)
print(d)
```
