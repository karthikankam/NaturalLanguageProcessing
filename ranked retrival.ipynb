```python
import re
def prepro(data):
    data=re.sub('[^a-z]+',' ',data)
    data=re.sub('[\s]+'," ",data)
    data=re.sub('-',' ',data)
    d=nlp(data)
    tok=[t.lemma_ for t in d if (t.is_stop==False)]
    return tok
    
```

```python
doc['tok']=doc['body'].apply(prepro)
doc['tok']
```

```python

def jaccard(d,q):
    num=len(set(d).intersection(set(q)))
    den=len(set(d).union(set(q)))
    jc=num/den
    return jc
    
```

```python
p=doc[['tok','docid']].copy()
p
```

```python
def jac(q):
    p['rel']=p['tok'].apply(lambda x:jaccard(x,q))
    rel=p.sort_values('rel',ascending=False).head()['docid'].values
    return rel
```

```python
qtr['jl']=qtr['tok'].apply(lambda x:jac(x))
qtr
```

```python
import numpy as np
qtr['ground']=qtr['qid'].apply(lambda x:qrel[qrel['qid']==x]['docid'].values)
qtr.sample(10).reset_index(drop=True)
```

```python
def avg_pres(mod,gro):
    t=0
    pres=[]
    for ind,val in enumerate(mod):
        if val in gro:
            t+=1
            pres.append(t/(ind+1))
    if len(pres)==0:
        return 0
    return np.mean(pres)
    
```

```python
qtr['jc']=qtr.apply(lambda x:avg_pres(x['jl'],x['ground']),axis=1)
qtr
```

```python
def tfidf(q):
    s=list(set(vocab).intersection(set(q)))
    col=['docid']
    col.extend(s)
    temp=dop.loc[:,col].copy()
    #tee=dop[:,col].copy()
    for i in s:
        for j in temp['docid'].values:
            temp.loc[temp['docid']==j,i]*=di[i]
    temp['tfidf']=temp[s].sum(axis=1)
    temp.sort_values('tfidf',ascending=False,inplace=True)
    g=temp.head()['docid'].values
    return g
```
