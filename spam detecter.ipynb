```python
import spacy
nlp=spacy.load('en_core_web_sm')
nlp.Defaults.stop_words
```

```python
def cleaned(data):
    doc=nlp(data)
    t=[tok.text for tok in doc if tok.is_stop==False]
    return " ".join(t)
```

```python
data['cleaned']=data['cleaned'].apply(cleaned)
```

```python
data['word count']=data['pure'].apply(lambda x:len(x.split()))
data['wordsintext']=data['text'].apply(lambda x:len(x.split()))
data['digits']=data['pure'].apply(lambda x :  sum([1 if w.isdigit() else 0 for w in x.split()]))
```

```python
data['with spaces']=data['pure'].apply(lambda x:abs(len(x.split())*2-1))
```

```python
data.drop('cleaned',inplace=True,axis=1)
```

```python
pos_dic = {"noun" : ["NNP", "NN", "NNS", "NNPS"], "verb" : ["VBZ", "VB", "VBD","VBG", "VBN"]}
```

```python
data.digits.unique()
```

```python
def puct(data,pos):
    doc=nlp(data)
    tok=[t for t in doc]
    count=0
    for i in tok:
        if i.tag_ in pos_dic[pos]:
            count+=1
    return count
```

```python
data['verb']=data['pure'].apply(lambda x:puct(x,'verb'))
data['noun']=data['pure'].apply(lambda x:puct(x,'noun'))
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

tar=LabelEncoder().fit_transform(data.label.values)
```

```python
tr=data[['word count','wordsintext','digits','verb','noun']]
```

```python
from sklearn.model_selection import train_test_split as tts
xtr,xts,ytr,yts=tts(tr,tar,test_size=0.2,random_state=9,stratify=tar)
```

```python
p1=mod.predict(xtr)
p2=mod.predict(xts)
```
